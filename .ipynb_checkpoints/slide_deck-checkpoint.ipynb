## Investigation Overview

In this investigation, I wanted to look at the characteristics of video games that could be used to increae their sales. The main focus was on the three Cs of video games: Platform,Genre,and Publisher.

## Dataset Overview

The data consisted of sales and attributes of approximately 16,000 round-cut video games. The attributes included the three Cs of video games, as well as additional attributes such as Na sales, Na sales, and Na sales. Thirty-nine data points were removed from the analysis due to inconsistencies or missing information.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
games=pd.read_csv('video_games.csv')
df=games[(games['Platform'] == 'PS4') |(games['Platform'] == 'PS2')|(games['Platform'] == 'PC')|(games['Platform'] == 'Wii')|(games['Platform'] == 'PS')|(games['Platform'] == 'DS')|(games['Platform'] == 'X360')|(games['Platform'] == 'PS3')] 
dfx=df[(df['Publisher'] == 'Nintendo')|(df['Publisher'] == 'Electronic Arts')|(df['Publisher'] == 'Activision')|(df['Publisher'] == 'Sony Computer Entertainment')|(games['Publisher'] == 'Ubisoft')]


```

## (increase sales around years)

We find that the video games industry is on the increase, and there are many companies entering that industry, and with the passage of years we notice an increase in the number of games produced, which leads to an increase in the purchase because it has become a profitable industry,and this is evident on the graph, which shows that sales are increasing every year

```python
year=games.groupby('Year')['Global_Sales'].sum().reset_index()
plt.figure(figsize=[14.7,8.27])
sb.lineplot(data=year,x='Year',y='Global_Sales');
plt.xlim((0,30))
plt.xticks(rotation=90);
plt.xlabel("Year");
plt.ylabel("Global sales");
plt.title('increase sales around years');
```

## (Publisher and sales)

We have talked that many companies enter the gaming industry, and here we find that there is great competition among the most important publishers in the sales of these video games.

```python
sb.catplot(data =dfx, y = 'Publisher', x = 'Global_Sales',kind="boxen",palette="Blues",orient='h',height=8.27,aspect=14.7/8.27)
plt.xlim((0,15))
plt.xticks(rotation=90);
plt.title("know sales for popular publisher");
plt.xlabel("Publisher");
plt.ylabel("Global sales");
```

## (Publisher vs Genre)
 We find that there are some companies that produce a certain type of games or produce that type in large numbers for many reasons, including that it is the most common type or the type that the publisher is able to produce something good for customers,like publisher Electronic arts ,activision and ubisoft there fouce on specific type of video games.



```python
ct_counts = dfx.groupby(['Publisher', 'Genre']).size()

# Use Series.reset_index() to convert a series into a dataframe object
ct_counts = ct_counts.reset_index(name='count')

# Use DataFrame.pivot() to rearrange the data, to have vehicle class on rows
ct_counts = ct_counts.pivot(index ='Genre' , columns = 'Publisher', values = 'count')
plt.figure(figsize=[14.7,8.27])
sb.heatmap(ct_counts, annot = True,cmap = 'mako_r',fmt='g');
plt.xlabel("Publisher");
plt.ylabel("Genre");
plt.title("Publisher and Genre count");
```

## (Platform vs Global sales)

ps4 has the most sales until now.


```python
sb.boxplot(data =dfx, y = 'Platform', x = 'Global_Sales',palette="Blues");
plt.xlim((0,5))
plt.ylabel("Platform");
plt.xlabel("Global sales");
plt.title("the most platform sales");
```
